```python
from jinja2 import Template, FileSystemLoader, Environment
import os

CWD = os.getcwd()
template_path = os.path.join(CWD,'templates')
blocks_path = os.path.join(template_path,'blocks')
def write_file(text):
    with open(os.path.join(template_path,'test-tmplt.html'),"w+") as f:
        f.write(text)
                
def get_jinja_template(file):
    jinja_folder = 'blocks/{}'.format(file)
    return os.path.join(template_path,jinja_folder)

def get_css():
    css_dir = os.path.join(template_path,'src/css')
    css = os.listdir(css_dir)
    return css

def get_js():
    js_dir = os.path.join(template_path,'src/js')
    js = os.listdir(js_dir)
    return js
```

```python
html_template = """
<!DOCTYPE html>
<html lang="en">
<head>
    <title>My Webpage</title>
</head>
<body>
    <ul id="navigation">
    {% for item in navigation %}
        <li><a href="{{ item.href }}">{{ item.caption }}</a></li>
    {% endfor %}
    </ul>

    <h1>My Webpage</h1>
    {{ a_variable }}

    {# a comment #}
</body>
</html>
"""
```

```python
def sort_js(js_lists=list()):
    indices = ['jquery','bootstrap']
    sorted_scripts = list()
    while js_lists:
        for index in indices:
            for js in js_lists:
                if index in js:
                    sorted_scripts.insert(len(sorted_scripts),js)
                    js_lists.remove(js)
                    break
        sorted_scripts.extend(js_lists)
        js_lists = list()
    return sorted_scripts
```

```python
sort_js(get_js())
```

```python
_template = Template(html_template)
file_loader = FileSystemLoader(blocks_path)
env = Environment(loader=file_loader)
template = env.get_template('child-content.jinja')
```

```python
stylesheets = [{"rel":"stylesheet","href":i} for i in get_css()]
js_scripts = get_js()
test = {"name": "test-template.html","success": "100"}

test_reports = [
    {
        "method": "GET",
        "group": "TESTGET",
        "status": "Success",
        "url": "www.google.com",
        "test_name": "GET_ALL",
        "is_failed": False
    },
    {
        "method": "GET_FAILED",
        "group": "TESTGET",
        "status": "Failed",
        "url": "www.facebook.com",
        "test_name": "GET_ALL",
        "is_failed": True
    },
    {
        "method": "GET_3",
        "group": "TESTGET",
        "status": "Success",
        "url": "www.googles.com",
        "test_name": "GET_None",
        "is_failed": False
    }]

class MyReports(object):
    
    def __init__(self,reports):
        for key,val in reports.items():
            setattr(self,key,val)

test_reports = [MyReports(i) for i in test_reports]
text = template.render(stylesheet=stylesheets,scripts=get_js(),name="Riszky",test=test,test_reports=test_reports)
write_file(text)
```
